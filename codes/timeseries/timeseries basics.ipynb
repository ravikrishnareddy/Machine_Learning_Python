```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For QQ-plot
from scipy.stats import probplot
import pylab

%matplotlib inline
```

```python
df_index_data_original = pd.read_csv(filepath_or_buffer='https://raw.githubusercontent.com/vishnukanduri/Time-series-analysis-in-Python/master/Index2018.csv', parse_dates=['date'], dayfirst=True)
df_index_data = df_index_data_original.copy()
df_index_data.head()
```

```python
df_index_data['spx'].plot(figsize=(15, 5), title="S&P500 Prices")
plt.show()
```

```python
df_index_data['ftse'].plot(figsize=(15, 5), title="FTSE Prices")
plt.show()
```

```python
df_index_data['spx'].plot(figsize=(15, 5), title="S&P500 Prices")
df_index_data['ftse'].plot(figsize=(15, 5), title="FTSE Prices")
plt.title("S&P500 vs. FTSE")
plt.show()
```

```python
# QQ-Plot of spx
probplot(df_index_data['spx'], plot=pylab)
plt.show()
```

```python
df_index_data['date'].describe(datetime_is_numeric=True)
```

```python
# Set 'data' column as index of the data
df_index_data.set_index(keys='date', inplace=True)
df_index_data.head()
```

```python
# Set desired frequency
df_index_data = df_index_data.asfreq('b')
df_index_data.head()
```

```python
# Handling missing values
df_index_data.isna().sum()
```

```python
# fill null values using forward fill
df_index_data.apply(lambda x:x.fillna(method='ffill', inplace=True))
df_index_data.isna().sum()
```

```python
# Let's create a proxy column 'index_price' and start analyzing spx data first
df_index_data['index_price'] = df_index_data['spx']
df_index_data.head()
```

```python
# split data into train and test datasets
train_size = int(len(df_index_data) * 0.8)

df_train = df_index_data.iloc[:train_size]
df_test = df_index_data.iloc[train_size:]

print(df_train.tail())
print(df_test.head())
```
